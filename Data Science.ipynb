```python
class Car(ABC):
    def __init__(self,company,price,model):
        self.compamy=company
        self.price=price
        self.model=model
        
    @abstractmethod
    def details(self):
        pass
    def speed(self):
        print("the speed is____")

class SUV(Car):
    def details(self):
        print("company name is", self.compamy)
        print("price is", self.price)
        print("model is", self.model)

class Cedan(Car):
     def detailsNA(self):
        print("company rating are not good", self.compamy)
        print("price is high", self.price)
        print("model is not good", self.model)
                  
s=SUV('abc',2400000,'xyz')
s.details()
c=Cedan('asd',10000000,'jkl')
c.detailsNA()

                                                                                                                                                                                                                                                                                                                                                                  
```
